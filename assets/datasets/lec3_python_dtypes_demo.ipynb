```python
import random 

vals = [1e-10, 1e-5, 1e-2, 1]
vals = vals + [-v for v in vals]

print(vals)
results = []
random.seed(12)
for _ in range(10000):
    random.shuffle(vals)
    results.append(sum(vals))

results = sorted(set(results))
print(f"There are {len(results)} unique results") 
```
